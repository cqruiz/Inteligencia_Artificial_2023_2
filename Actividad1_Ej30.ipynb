```python
import numpy as np
import matplotlib.pyplot as plt

KL_D1_Mu = []
KL_D1_Var = []
KL_D2_Mu = []
KL_D2_Var = []
rango = np.linspace(0.1,3,100)

for x in rango:
    s = 1
    sigma = 1
    mu = 0
    m = x
    KL_D1_Mu.append(np.log(s/sigma)+1/(2*(s/sigma)**2)-0.5+((mu-m)/(2*s))**2)
    s = 1
    sigma = 1
    mu = x
    m = 0
    KL_D2_Mu.append(np.log(s/sigma)+1/(2*(s/sigma)**2)-0.5+((mu-m)/(2*s))**2)

for x in rango:
    s = x
    sigma = 1
    mu = 0
    m = 0
    KL_D1_Var.append(np.log(s/sigma)+1/(2*(s/sigma)**2)-0.5+((mu-m)/(2*s))**2)
    s = 1
    sigma = x
    mu = 0
    m = 0
    KL_D2_Var.append(np.log(s/sigma)+1/(2*(s/sigma)**2)-0.5+((mu-m)/(2*s))**2)
```

```python
plt.plot(rango, KL_D1_Mu, label='Primera distribución normalizada')
plt.plot(rango, KL_D2_Mu, 'r--',label='Segunda distribución normalizada')
plt.title('KL Divergence respecto a la media')
plt.xlabel('Media')
plt.ylabel('KL Divergence')
plt.legend()
```

```python
plt.plot(rango[10:], KL_D1_Var[10:], label='Primera distribución normalizada')
plt.plot(rango[10:], KL_D2_Var[10:], 'r--', label='Segunda distribución normalizada')
plt.title('KL Divergence respecto a la Varianza')
plt.xlabel('Varianza')
plt.ylabel('KL Divergence')
plt.legend()
```
